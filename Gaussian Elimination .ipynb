```python
# Write a program to implement the Gaussian elimination method

A = [[2, 1, 1], [3, 2, 3], [1, 4, 9]]
b = [10, 18, 16]

# Forward Elimination
n = len(A)
for i in range(n):
    for j in range(i+1, n):
        factor = A[j][i] / A[i][i]
        for k in range(i, n):
            A[j][k] -= factor * A[i][k]
        b[j] -= factor * b[i]

# Back Substitution
x = [0] * n
for i in range(n-1, -1, -1):
    x[i] = b[i]
    for j in range(i+1, n):
        x[i] -= A[i][j] * x[j]
    x[i] /= A[i][i]

print("Solution:")
print(x)
```

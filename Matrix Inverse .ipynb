```python
# Write a program to enter a matrix and check if it is invertible. If the inverse exists, find the inverse

n = int(input("Enter the size of rows of the matrix :"))
print("\nEnter the elements of the matrix row-wise :")
matrix = []
for i in range(n):
    row = list(map(float,input().split()))
    matrix.append(row)
    
if np.linalg.det(matrix)==0:
    print("Matrix is not invertible")
else:
    inverse = np.linalg.inv(matrix)
    print("\nInverse of the matrix :")
    for row in inverse:
        for element in row:
            print(round(element,2),end =" ")
        print()
```

```python
# Plot set of complex nos

P = {4+1j, 1+5j , 2-7j, 2.5+3j, 3.27+6j}

import matplotlib.pyplot as plt
def plot(P):
    n = {x for x in P}
    X = [x.real for x in n]
    Y = [x.imag for x in n]
    plt.plot(X,Y,'ro')

plot(P)
```

```python
# new plot by rotating the given number by a degree 90, 180, 270 degrees and also by scaling by a number a=1/2, a=1/3, a=2 

angle = int(input("Enter Angle of Rotation :"))
if angle==90:
    r = {x * 1j for x in P}
    print("P after 90 degree Rotation :",r)
    plot(r)
elif angle==180:
    r = {x * (-1) for x in P}
    print("P after 180 degree Rotation :",r)
    plot(r)
elif angle==270:
    r = { x * 1j *(-1) for x in P}
    print("P after 270 degree Rotation :",r)
    plot(r)
else:
    print("Invalid Angle entered")

scale = float(input("Enter Scale point :"))
r = {x * scale for x in P}
print("Scale :",r)
plot(r)
```
